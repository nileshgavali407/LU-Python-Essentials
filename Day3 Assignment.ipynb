# Question 1: Report Card using IF ELIF

```python
i=1
marks=70
while i<=1:
  if marks >=85:
         print("| -----Grade A----- |")
         i +=1
  elif marks <85 and marks >=60:    
         print("| -----Grade B----- |")
         i +=1
  elif marks <60 and marks >=50:
         print("| -----Grade C----- |")
         i +=1
  elif marks <50 and marks >=40:
         print("| -----Grade D----- |")
         i +=1
  else:
         print("|     FAIL   |")
         i +=1
```

# Question 2: Prime Numbers from 1-1000 using For loop

```python
#Take the input from the user:   
lower = int(input("Enter lower range: "))  
upper = int(input("Enter upper range: "))  
  
for num in range(lower,upper + 1):  
   if num > 1:  
       for i in range(2,num):  
           if (num % i) == 0:  
               break  
       else:  
           print(num)  
        
```

# Question 3: Multiplication table 

```python
# Python Program to Print Multiplication Table of a Number

num = int(input("Enter the number: "))

print("Multiplication Table of", num)
for i in range(1, 11):
   print(num,"X",i,"=",num * i)
```
